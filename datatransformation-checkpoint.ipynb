```python
# concat and append
df1=df[['id','order_id','order_date','ship_date']]
df2=df[['id','product_id','product_name','category']]
df3=df[['id','quantity','sales','profit']]
df4=df[['id','region','states','city']]
frames=[df1,df2,df3,df4]
rslt=pd.concat(frames)
#rslt=df1.append(frames)
#rslt['category'].unique()
rslt
```

```python
# merge
result = pd.merge(df1, df2, on='id')
result
```

```python
result1 = pd.merge(result, df3, on='id')
result1
```

```python
#Joining / merging on duplicate keys can cause a returned frame that is the multiplication of the row
#dimensions, which may result in memory overflow. It is the user’ s responsibility to manage duplicate values inn 
#keys before joining large DataFrames

result2 = pd.merge(result1, df4, how='inner',on='id',validate="one_to_one")
result2
```

```python

result3 = df1.join(df3,on='id', how='inner', lsuffix='_left', rsuffix='_right')
result3
```
